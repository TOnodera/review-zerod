```python
import numpy as np
from common.functions import softmax, cross_entropy_error

# 乗算レイヤー
class MulLayer:
    def __init__(self):
        self.x = None
        self.y = None
        
    def forward(self, x, y):
        # 逆伝播で使う値を保持しておく
        self.x = x
        self.y = y
        return x*y
    
    def backward(self, dout):
        # 乗算の逆伝搬は値をひっくり返す処理
        dx = dout * self.y
        dy = dout * self.x

        return dx, dy
    
# 加算レイヤー
class AddLayer:
    def __init__(self):
        pass

    def forward(self, x, y):
        return x+y
    
    def backward(self, dout):
        # 加算の場合は特に何もしない
        return dout, dout
    

# Reluレイヤー
class Relu:
    def __init__(self):
        self.mask = None
        
    def forward(self, x):
        self.mask = x<=0
        out = x.copy()
        out[self.mask] = 0

        return out

    def backward(self, dout):
        # Trueのところ(x<=0)を0にする
        # それ以外の値はそのまま保持
        dout[self.mask] = 0
        return dout
    
# シグモイド関数
class Sigmoid:
    def __init__(self):
        self.out = None
        
    def forward(self, x):
        out = 1/(1+np.exp(-x))
        self.out = out
        return out
    
    def backward(self, dout):
        return dout * (1.0-self.out) * self.out



# アフィン変換
# https://qiita.com/masatomix/items/0678e56e56af27efa110
# ↑この解説がわかりやすい
class Affine:
    def __init__(self, W, b):
        self.W = W
        self.b = b
        self.x = None
        self.dW = None
        self.db = None
        
    def forward(self, x):
        self.x = x
        return np.dot(x, self.W) + self.b
    
    def backward(self, dout):
        dx = np.dot(dout, self.W.T)
        self.dW = np.dot(self.x.T, dout)
        self.db = np.sum(dout, axis=0)
        return dx
    
# softmax関数
class SoftmaxWithLoss:
    def __init__(self):
        self.loss = None
        self.y = None
        self.t = None
        
    def forward(self, x, t):
        self.t = t
        self.y = softmax(x)
        self.loss = cross_entropy_error(self.y, self.t)
        
        return self.loss
    
    def backward(self, dout=1):
        batch_size = self.t.shape[0]
        return (self.y - self.t) / batch_size
        
x = np.array([1,2,3])
y = np.array([2,3,4])

(x-y)/3







```
